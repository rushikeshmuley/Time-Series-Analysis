```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
from pylab import rcParams
rcParams['figure.figsize']=14,7
decomposotion_additive=seasonal_decompose(df,model='additive')
decomposotion_multiplicative=seasonal_decompose(df,model='multiplicative')
```

```python
decomposotion_additive.plot()
plt.show()
```

```python
decomposotion_multiplicative.plot()
plt.show()
```

```python
# Check Autocorrelation-Durbin Watson Test
```

```python
import statsmodels.api as sm
```

```python
sm.stats.durbin_watson(df)
```

```python
# Durbin value less than 2 means positive autocorrelation
```

```python
# The Durbin Watson (DW) statistic is a test for autocorrelation in the residuals from a statistical model or regression analysis. The Durbin-Watson statistic will always have a value ranging between 0 and 4. A value of 2.0 indicates there is no autocorrelation detected in the sample.
```
