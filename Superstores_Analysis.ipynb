```python
df1.sort_values('Order Date',inplace=True)
df1.head()
```

```python
df1=df1[["Order Date","Sales"]]
df1.head()
```

```python
df1=df1.groupby("Order Date")["Sales"].sum().reset_index()
```

```python
df1
```

```python
df1=df1.set_index("Order Date")
df1.head()
```

```python
y=df1["Sales"].resample("MS").mean()
```

```python
y["2014"]
```

```python
y["2015"]
```

```python
y.plot(figsize=(15,8))
```

```python
import matplotlib.pyplot as plt
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decompose=seasonal_decompose(y)
decompose.plot()
plt.show()
```

```python
# Durbin watson Test
import statsmodels.api as sm
sm.stats.durbin_watson(y)
```
